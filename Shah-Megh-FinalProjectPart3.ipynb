## Club Players Analysis

### Here I am performing analysis on Work Rate, Club Position, Age, and Overall Rating.

### There are 9 unique values for work rate. Work Rate refers to the work a player does by running and chasing a ball when the ball is not in possession. Work Rate is an important factor in football and hence we are considering this attribute.
### Club Position means the position at which player plays for his club.
### Age is the age of the player in years.
### Overall Rating means how good is the player on a scale of 100.

Now I am creating a pivot table for the club position data. Here I am taking the index as the work rate, the columns are the various positions that players play at a club. The value of the heat map is the age. The aggregate function which I am using here is mean. We will fifa_club_pivot table and use this for making our first heat map. The barplot is used for club data. This barplot has age as values on the X-axis and the Overall Rating of the player on the Y-axis. I am using bqplot to make this barplot. This is a vertical bar chart. I am also setting the marks, axes, and margins for the bar plot in the code written below. Next, I am creating a heat map for the position of player w.r.t club. We are using the fifa_club_pivot table here. First, we define the Scale for our heatmap. Here we define both as ordinal scale as club position and work rate are ordinal data. Next, we define the axis. Here we have Club Position as the X-axis and Work Rate as the Y-axis. These are the names visible on the heat map. Next, we define the marks. Here we mention specific characteristics for our heat map. We define the color to be from the values of the fifa_club_pivot table. Their rows are the index and columns are columns of the pivot table. We define the scales from the Scales defined earlier. As it is an interactive heat map, the interaction is defined as select. The anchor style, i.e. when we select something on the heat is set to black. We then write a piece of code to define the interactions with our barplot. When we select a block on the heat_map, the corresponding value of age and overall rating is shown on the bar plot. We have shown the mean of overall_rating in the barplot. We use ipywidgets.Label() function to print something when a block of the heatmap is selected. We define the on_selected_club function when a block of the heatmap for club data is selected. We print the Mean of Age for that particular work rate of that particular position. We finally define the map type as heat_map in the marks of our figure. We then print a dashboard. This is a vertical dashboard that comprises the heat map and bar plot.

```python
#Pivot table for club data
fifa_club_pivot = pd.pivot_table(fifa, index = 'work_rate', columns = 'club_position', values = 'age', aggfunc = 'mean', fill_value=0)

#Barplot for Club
x_scs = bqplot.OrdinalScale()
y_scs = bqplot.LinearScale()

x_axs = bqplot.Axis(label='Age', scale=x_scs)
y_axs = bqplot.Axis(label='overall Rating', scale=y_scs, 
                   orientation = 'vertical')

bar_plot1 = bqplot.pyplot.bar(x = [], 
                         y = [],
                         scales={'x':x_scs, 'y':y_scs})

fig_dur1 = bqplot.Figure(marks=[bar_plot1], axes=[x_axs,y_axs], fig_margin={'top':20, 'bottom':60, 'left':40, 'right':80})

# Code for the Heat Map
# Scales
col_sc = bqplot.ColorScale(scheme='Red', min=np.nanmin(fifa_club_pivot.values), max=np.nanmax(fifa_club_pivot.values))
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale() 

# Axis
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
x_ax = bqplot.Axis(scale=x_sc, label = 'Club Position')
y_ax = bqplot.Axis(scale=y_sc, orientation='vertical', label = 'Work Rate')

# Marks
heat_map = bqplot.GridHeatMap(color=fifa_club_pivot.values, 
                             row = fifa_club_pivot.index, column = fifa_club_pivot.columns, 
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, 
                             interactions={'click':'select'}, 
                             anchor_style={'fill':'black'})

def barplot_fig1(x, y):
    df2 = fifa[(fifa['club_position'] == y) & (fifa['work_rate'] == x)]
    df2 = df2[df2['age'] != 0]
    df2 = df2.groupby('age')['overall_rating'].mean()
    return df2

mySelectedLabel1 = ipywidgets.Label()
def on_selected_club(change):
    if len(change['owner'].selected) == 1: # only allow user to select 1 bin
        i,j = change['owner'].selected[0]
        v = fifa_club_pivot.values[i,j] # data value in that bin
        mySelectedLabel1.value = 'Mean Age for Work Rate: ' + fifa_club_pivot.index[i] + ', and Position: ' + fifa_club_pivot.columns[j] + ' = ' + str(v)
        fifa_club_pivot2 = barplot_fig1(fifa_club_pivot.index[i], fifa_club_pivot.columns[j])
        bar_plot1.x = fifa_club_pivot2.index
        bar_plot1.y = fifa_club_pivot2.values
heat_map.observe(on_selected_club, 'selected')

# Figure
fig1 = bqplot.Figure(marks=[heat_map], axes=[col_ax, x_ax, y_ax], fig_margin={'top':30, 'bottom':60, 'left':80, 'right':100})
```

## How to Use the Club Player Analysis Dashboard?

### In the dashboard given below, you just need to select a block of a heatmap that corresponds to a playing position(national or club) and work rate. You can see the mean age for that position and work rate. Below, in the bar chart, you can see the age groups playing at that position and the mean of their overall ratings. You just have to click on a particular block of the heatmap and the mean value of age for that work rate at that particular position is displayed. 

### For example, in Club Position Analysis Dashboard if the you select the block where "CAM" is Club Position for "High/High" Work Rate, we can see that the mean age value is 26.6 years. The barplot also displays the various age values and mean of overall rate for that "CAM" club position and "High/High" Workrate.

```python
myDashboard1 = ipywidgets.VBox([mySelectedLabel1, fig1, fig_dur1])
myDashboard1
```

## National Player Analysis

### We will perform National Player Analysis with respect to factors such as Work Rate, Age, Overall Rating, and National Position.

### This is similar to club player analysis. Here National Position means position at which player plays for their nation.


Just like club data, now we will create a dashboard for the national team data. First, we will create a pivot table for the national team. Here again, we are using the position of the player, but instead of a club, it is for a national team. Other factors such as age and work rate remain the same just like the first pivot table. I am writing my code for the barplot. Just like club player analysis over here too we have age and the overall rating on the X-axis and Y-axis respectively. I have created code for the heatmap and dashboard of the national team position. This is similar to what we did for the club player analysis dashboard. Here we are showing the mean age for that particular work rate at that position.

```python
# Code for Pivot Table
fifa_national_pivot = pd.pivot_table(fifa, index = 'work_rate', columns = 'national_team_position', values = 'age', aggfunc = 'mean', fill_value=0)

# Barplot for National Team
x_scs = bqplot.OrdinalScale()
y_scs = bqplot.LinearScale()

x_axs = bqplot.Axis(label='Age', scale=x_scs)
y_axs = bqplot.Axis(label='overall rating', scale=y_scs, 
                   orientation = 'vertical')

bar_plot2 = bqplot.pyplot.bar(x = [], 
                         y = [],
                         scales={'x':x_scs, 'y':y_scs})

fig_dur2 = bqplot.Figure(marks=[bar_plot2], axes=[x_axs,y_axs], fig_margin={'top':20, 'bottom':60, 'left':40, 'right':80})

#Heat Map Code for National Player Analysis
# Scales
col_sc = bqplot.ColorScale(scheme='Red', min=np.nanmin(fifa_national_pivot.values), max=np.nanmax(fifa_national_pivot.values))
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale() 

# Axis
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
x_ax = bqplot.Axis(scale=x_sc, label = 'National Team Position')
y_ax = bqplot.Axis(scale=y_sc, orientation='vertical', label = 'Work Rate')

# Marks
heat_map = bqplot.GridHeatMap(color=fifa_national_pivot.values, 
                             row = fifa_national_pivot.index, column = fifa_national_pivot.columns, 
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, 
                             interactions={'click':'select'}, 
                             anchor_style={'fill':'black'})

def barplot_fig2(x, y):
    df2 = fifa[(fifa['national_team_position'] == y) & (fifa['work_rate'] == x)]
    df2 = df2[df2['age'] != 0]
    df2 = df2.groupby('age')['overall_rating'].mean()
    return df2

mySelectedLabel2 = ipywidgets.Label()
def on_selected_national(change):
    if len(change['owner'].selected) == 1: # only allow user to select 1 bin
        i,j = change['owner'].selected[0]
        v = fifa_national_pivot.values[i,j] # data value in that bin
        mySelectedLabel2.value = 'Mean Age for Work Rate: ' + fifa_national_pivot.index[i] + ', and Position: ' + fifa_national_pivot.columns[j] + ' = ' + str(v)
        fifa_national_pivot2 = barplot_fig2(fifa_national_pivot.index[i], fifa_national_pivot.columns[j])
        bar_plot2.x = fifa_national_pivot2.index
        bar_plot2.y = fifa_national_pivot2.values
heat_map.observe(on_selected_national, 'selected')

# Figure
fig2 = bqplot.Figure(marks=[heat_map], axes=[col_ax, x_ax, y_ax], fig_margin={'top':30, 'bottom':60, 'left':80, 'right':100})
```

## How to Use the National Player Analysis Dashboard?

### Similar to the Club Player Analysis Dashboard, you can use the national player analysis dashboard by selecting a particular block of the heatmap. The heatmap is for the national team position. The positions which players play for a domestic club can be different from an international team, and hence a separate dashboard to analyze this. For example, in the second dashboard, if a user selects the block where the National Team position is "CAM" and work rate is "High/High", we get mean as 27.86 years. This average value differs from that of the club position. The reason is the position for the same player may be different when he plays for the national team. We see corresponding differences in the barplot as well.

```python
myDashboard2 = ipywidgets.VBox([mySelectedLabel2, fig2, fig_dur2])
myDashboard2
```
