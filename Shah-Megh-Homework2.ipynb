Now I will try to plot the histogram for the red channel of the animated image. Here the fig size is given as 10 x 5. For the ax.hist function, we take all the columns, all the rows of the dog_animated_data. Here 0 represents the Red color channel.

```python
dog_animated_data[:,:,0].shape
```

Above we have found the shape for the data of our histogram plot. We see that this is a 2D dataset. This was what we had inputted in the histogram earlier. That is why we get different colors for different columns. We don't want the rows from where the data is coming from hence, we need to mutate our data to a 1D array. For mutating to 1D array we will use the flatten() function.

```python
dog_animated_data[:,:,0].flatten()
```

Now, I have unwrapped the data. We can see that below is the original 2D array. We have mutated this into a 1D array.

```python
dog_animated_data[:,:,0]
```

```python
dog_animated_data[:,:,0].flatten().shape
```

```python
590*590
```

We see that the 1D array is taking 590x590, which is equal to 348100, just like the shape of over mutated array. Hence, we have unwrapped the whole thing. Now we will make the histogram for this 1D array.

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(dog_animated_data[:,:,0].flatten())
plt.show()
```

The color range in each channel ranges from 0-255. Hence there are total 256 bins for each color channel. Below I am plotting the R, G and B channel bins against the number of pixels in each bin.

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(dog_animated_data[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, 0th or first channel = Red channel
ax.hist(dog_animated_data[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax.hist(dog_animated_data[:,:,2].flatten(), bins=256, color='blue') # Blue Channel
ax.set_xlabel('R & G & B channel bins')
ax.set_ylabel('Number of pixels in each bin')
plt.show()
```

I can see that this histogram is quite hard to see, it is not visually pleasing. I see that there is some overlapping happening as well for the different Red, Green and Blue channels. To avoid these issues, we need to make separate axis axes for each color channel. We will make a series of plots, i.e. 3 different plots for each color channel. So now we will make 3 rows of plots and 1 column of the plot.

```python
fig, ax = plt.subplots(4,1,figsize=(15,15)) #Here 4 represents 3 for the R,G, and B channels and 1 for the plot of the imahge
ax[0].hist(dog_animated_data[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, Red channel
ax[1].hist(dog_animated_data[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax[2].hist(dog_animated_data[:,:,2].flatten(), bins=256, color='blue') # Blue Channel
ax[0].set_xlabel('R channel bins') #X-axis label for the Red Channel Bins
ax[1].set_xlabel('G channel bins') #X-axis label for the Green Channel Bins
ax[2].set_xlabel('B channel bins') #X-axis label for the Blue Channel Bins
ax[0].set_ylabel('Number of pixels') #Y-axis label for the number of pixels
ax[1].set_ylabel('Number of pixels') #Y-axis label for the number of pixels
ax[2].set_ylabel('Number of pixels') #Y-axis label for the number of pixels
ax[3].imshow(dog_animated_data) #This includes the animated dog image in the end
plt.show()
```

The peak R-G-B values of this histogram are at 255-255-255. Hence this image is on the brighter side. We can clearly see that the background of the image is white and most parts of the dog is white in color, hence the high values around 255 is justifiable.
